# importing required libraries

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder,LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
X=new_df.drop(new_df[["salary"]],axis=1)
y=new_df[["salary"]]
```

```python
X=X.drop(columns=['sex','race','capital-loss','country'],axis=1)
```

```python
category_cols=X.select_dtypes(include=['object']).columns
category_cols
```

```python
for i in category_cols:
    print(f"Unique value of {i} :")
    print(new_df[i].unique())
    print()
```

```python
numeric_cols=X.select_dtypes(exclude=['object']).columns
numeric_cols
```

```python
# l = LabelEncoder()

# df_cat = new_df[category_cols]
# for i in df_cat.columns:
#     if df_cat[i].dtypes=="object":
#         df_cat[i]=l.fit_transform(df[i])
```

```python
num_pipeline=Pipeline(
    steps=[("imputer",SimpleImputer()),
           ("scaler",StandardScaler())]
)
```

```python
cat_pipeline = Pipeline(
    steps=[("imputer", SimpleImputer(strategy="most_frequent")),
           ("encoder", OneHotEncoder(handle_unknown="ignore"))]
)
```

```python
preprocessor=ColumnTransformer(
    [
        ("num_pipeline",num_pipeline,numeric_cols),
        ("cat_pipeline",cat_pipeline,category_cols) 
    ]
)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=40)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
type(X_train)
```

```python
train_data=preprocessor.fit_transform(X_train)
```

```python
type(train_data)
```

```python
# Get feature names from the ColumnTransformer
feature_names = preprocessor.named_transformers_['num_pipeline'].named_steps['scaler'].get_feature_names_out(numeric_cols).tolist() + \
                preprocessor.named_transformers_['cat_pipeline'].named_steps['encoder'].get_feature_names_out(category_cols).tolist()
```

```python
test_data=preprocessor.transform(X_test)
test_data.shape
```

```python
dense_train_data = pd.DataFrame(train_data.toarray(), columns=feature_names)
dense_test_data = pd.DataFrame(test_data.toarray(), columns=feature_names)
dense_train_data.head()
```

```python
#preprocessor.get_feature_names_out()
cols=dense_train_data.columns
```

```python
X_train=pd.DataFrame(dense_train_data,columns=cols)
X_test=pd.DataFrame(dense_test_data,columns=cols)
```

```python
X_train
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
models={
    "LogisticRegression":LogisticRegression(),
    "SVC":SVC(),
    "DecisionTreeClassifier":DecisionTreeClassifier()
}
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    print(model)
```

```python
def model_evaluate(test,pred):
    acc_score=accuracy_score(test,pred)
    conf_matrix=confusion_matrix(test,pred)
    class_report=classification_report(test,pred)
    return acc_score,conf_matrix,class_report
```

```python
for i in range(len(models)):
    model=list(models.values())[i]
    model.fit(X_train,y_train)
    
    y_pred=model.predict(X_test)
    acc_score,conf_matrix,class_report=model_evaluate(y_test,y_pred)
    
    print(list(models.keys())[i],">>>")
    print(f"Accuracy score : {acc_score}")
    print(f"Confusion matrix : \n {conf_matrix}")
    print(f"Classsification Report : \n {class_report}")
    print()
    print(50*"=")
```
